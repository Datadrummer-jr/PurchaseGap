Datapyme es un proyecto de ciencia de datos que busca iluminar el complejo panorama econmico de la Cuba contempornea a travs de la programacin y el anlisis de datos. Su objetivo principal es examinar cmo las fluctuaciones del dlar estadounidense y de otras divisas extranjeras influyen en las estrategias de precios de las MIPYMES (micro, pequeas y medianas empresas) y de las tiendas en MLC (moneda libremente convertible). Al comparar de manera sistemtica la dinmica de los precios locales con referentes globales, el proyecto revela patrones de inflacin, procesos de adaptacin empresarial y cambios estratgicos en la accesibilidad del consumidor.

Ms all del dlar, Datapyme ampla su alcance para analizar el comportamiento general de las divisas extranjeras en Cuba, explorando cmo interactan con las estructuras econmicas locales y cmo impactan en las transacciones cotidianas. Este enfoque multidimensional permite comprender con mayor profundidad la bsqueda de autonoma econmica, la resiliencia de las empresas y los retos que imponen las presiones del mercado global.

En ltima instancia, Datapyme no es solo un ejercicio tcnico de programacin y ciencia de datos, sino tambin una exploracin estratgica de cmo los conocimientos basados en datos pueden revelar los mecanismos ocultos de la supervivencia, la adaptacin y la independencia econmica.
