Datapyme es un proyecto de ciencia de datos que busca iluminar el complejo panorama econmico de la Cuba contempornea a travs de la programacin y el anlisis de datos. Su objetivo principal es examinar cmo las fluctuaciones del dlar estadounidense y de otras divisas extranjeras influyen en las estrategias de precios de los nuevos actores econmicos creados desde 2021. Al comparar de manera sistemtica la dinmica de los precios locales con referentes globales, el proyecto revela patrones de inflacin, procesos de adaptacin empresarial y cambios estratgicos en la accesibilidad del consumidor.

Ms all del dlar, Datapyme ampla su alcance para analizar el comportamiento general de las divisas extranjeras en Cuba, explorando cmo interactan con las estructuras econmicas locales y cmo impactan en las transacciones cotidianas. Este enfoque multidimensional permite comprender con mayor profundidad la bsqueda de autonoma econmica, la resiliencia de las empresas y los retos que imponen las presiones del mercado global.

En ltima instancia, Datapyme no es solo un ejercicio tcnico de programacin y ciencia de datos, sino tambin una exploracin estratgica de cmo los conocimientos basados en datos pueden revelar los mecanismos ocultos de la supervivencia, la adaptacin y la independencia econmica. Su propsito es aportar un recurso valioso para estudiantes, investigadores y emprendedores interesados en la interseccin entre tecnologa, economa y transformacin social.

```python
import plotly.graph_objects as go

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

fig = go.Figure()
fig.add_trace(go.Bar(
    x=months,
    y=[20, 14, 25, 16, 18, 22, 19, 15, 12, 16, 14, 17],
    name='Primary Product',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=months,
    y=[19, 14, 22, 14, 16, 19, 15, 14, 10, 12, 12, 16],
    name='Secondary Product',
    marker_color='lightsalmon'
))

# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_layout(barmode='group', xaxis_tickangle=-45, title="Grfica")
fig.show()
```

```python

from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=3, cols=1)

fig.add_trace(go.Scatter(
    x=[3, 4, 5],
    y=[1000, 1100, 1200],
), row=1, col=1)

fig.add_trace(go.pie
    x=[2, 3, 4],
    y=[100, 110, 120],
), row=2, col=1)

fig.add_trace(go.Scatter(
    x=[0, 1, 2],
    y=[10, 11, 12]
), row=3, col=1)


fig.update_layout(height=600, width=600, title="Stacked Subplots")
fig.show()
```
